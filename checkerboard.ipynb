```python
import numpy as np
import math
```

```python
import matplotlib.pyplot as plt
```

```python
N = 1000
```

```python
def checkerboard(n):
    x=np.random.rand(n,1)*5
    y=np.random.rand(n,1)*5
    x1 = [math.floor(i) for i in x]
    y1 = [math.floor(j) for j in y]
    c=np.remainder(np.add(x1,y1),2)
    return x,y,c
x,y,c=checkerboard(2000)
ind=np.where(c>0)
ind1=np.where(c==0)
plt.title('Checker data')
plt.plot(x[ind],y[ind],'.',label='Class 1')
plt.plot(x[ind1],y[ind1],'.',label='Class 2')
plt.show()
```

```python
bias = np.ones((2000,1))
```

Input X created 

```python
a1=np.hstack((x[ind],y[ind]))
a2=np.hstack((x[ind1],y[ind1]))
X1=np.vstack((a1,a2))
X=np.hstack((X1,bias))
```

```python
s1=np.shape(ind)
s1=s1[1]
s2=np.shape(ind1)
s2=s2[1]
```

```python
d1=np.zeros((s1,1))
d2=np.ones((s2,1))
d=np.vstack((d1,d2))
```

```python
from keras.models import Sequential
from keras. layers import Dense

model = Sequential()
model.add(Dense(32, input_dim=3, activation='tanh'))
model.add(Dense(90, input_dim=32, activation='tanh'))
model.add(Dense(100, input_dim=90, activation='tanh'))
model.add(Dense(100, input_dim=100, activation='tanh'))
model.add(Dense(100, input_dim=100, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(100, input_dim=150, activation='tanh'))
model.add(Dense(1, activation='tanh'))
model.compile(loss='mean_squared_error', optimizer = 'sgd', metrics = ['accuracy'])
model.fit(X,d,epochs=1000)
model.summary()
```

```python
y_pred = model.predict(X)
score = model.evaluate(X,d)
```

```python
accuracy = score[1]*100
print(accuracy)
```
